## 1. Tri par sélection

Rappel : fonction de recherche de la position du maximum dans une liste

```python
def recherche_max(tab):
    max = tab[0]
    pos_max = 0
    for i in range(1,len(tab)) :
        if tab[i] > max :
            max = tab[i]
            pos_max = i
    return pos_max
```

```python
def place_dans_liste(tab,val):
    i=0
    while val > tab[i] and i < len(tab) :
        i = i + 1
    return i
```

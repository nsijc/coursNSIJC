## Activité 2 : organiser une conférence

Objectif : Des conférenciers sont invités à présenter leurs exposés dans une salle. Mais leurs disponibilités ne leur permettent d’intervenir qu’à des horaires bien définis. Le problème est de construire un planning d’occupation de la salle avec le plus grand nombre de conférenciers.

Remarque : dans la suite du texte, les conférenciers seront désignés par les symboles C<sub>i</sub> où i est un entier naturel positif. Leurs horaires possibles seront donnés sous formes d'intervalles de type \[hd ; hf[ où hd et hf désignent respectivement les heures de début et de fin. 

### premier exemple :

Il y a 4 conférenciers avec les horaires suivants : C1 ∶\[3,4[ C2 ∶[0,1[ C3 ∶[2,3[ C4 ∶[1,2[ 

On peut visualiser les horaires sur un graphique grâce au code suivant :

```python
import matplotlib.pyplot as plt
## les données des conférenciers
conf=[[3,4],[0,1],[2,3],[1,2]]
## le graphique
def affiche_graphe(conf) :
    """ prend en arguments la liste conf des données des conférenciers
        affiche le graphique """
    # récupération de la taille des données à afficher
    nb_conferenciers = len(conf) # le nombre de conférenciers 
    horaires_max=[conf[i][1] for i in range(nb_conferenciers)] # les horaires de fin de chaque conférencier
    plage_horaire=max(horaires_max) # l'horaire maximum
    # gestion du graphique
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure()
    ax = plt.axes()
    ax = ax.set(xlabel='créneaux horaires', ylabel='conférenciers')
    plt.axis([0, plage_horaire+1, 0, nb_conferenciers+1]) #pour les graduations, de façon à voir les axes
    for i in range(len(conf)) :
        plt.plot(conf[i],[i+1,i+1]) # affichage des lignes conférencier par conférencier
    plt.show() #indispensable pour l'affichage
affiche_graphe(conf)
```
